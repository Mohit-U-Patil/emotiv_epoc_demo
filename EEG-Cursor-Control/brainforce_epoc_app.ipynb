```python
def quattorotmat(q):
    q00,q10,q20,q30 = normalize(q)
    #trying to match axis to cube
    #this replacement matches axis properly with implementation  rot_mat = angletorotmat(-y_rot, -x_rot, -z_rot) 
    q0 = q00
    q1 = q30
    q2 = q10
    q3 = q20

    #trial block: convert quat to euler angle, make rotation matrix to verify

    y_rot = math.asin( 2*(q0*q2 - q1*q3) )

    if y_rot == numpy.pi/2:
        x_rot = 0
        z_rot = -2*math.atan2(q1,q0)
    elif y_rot == -numpy.pi/2:
        x_rot = 0
        z_rot = 2*math.atan2(q1,q0)
    else:
        x_rot = math.atan2( 2*(q0*q1 + q2*q3), q0**2 - q1**2 - q2**2 + q3**2 )
        z_rot = math.atan2( 2*(q0*q3 + q1*q2), q0**2 + q1**2 - q2**2 - q3**2 )

    return(x_rot, y_rot, z_rot)
    
#this function convertes given rotation angles into rotation matrix
def angletorotmat4(ang_x, ang_y, ang_z):
    # import numpy as np
    # ang_x = ang_y = ang_z = np.pi/4
    r = R.from_euler('zyx', [ang_z, ang_y, ang_x], degrees=False)
    X = r.as_matrix()
    X0 = numpy.zeros((3,1))
    Xnew = numpy.hstack((X,X0))
    Xlast = numpy.array([0, 0, 0, 1] )
    Xfinal = numpy.vstack((Xnew,Xlast))        
    rot_mat_in = numpy.array(Xfinal,'f')
#     print(rot_mat_in)
    return(rot_mat_in)

def angletorotmat3(ang_x, ang_y, ang_z):
    # import numpy as np
    # ang_x = ang_y = ang_z = np.pi/4
    r = R.from_euler('zyx', [ang_z, ang_y, ang_x], degrees=False)
    X = r.as_matrix()
          
    rot_mat_in = numpy.array(X,'f')
#     print(rot_mat_in)
    return(rot_mat_in)

def normalize(v, tolerance=0.00001):
    mag2 = sum(n * n for n in v)
    if abs(mag2 - 1.0) > tolerance:
        mag = numpy.sqrt(mag2)
        v = tuple(n / mag for n in v)
    return v

def q_mult(q1, q2):
    w1, x1, y1, z1 = q1
    w2, x2, y2, z2 = q2
    w = w1 * w2 - x1 * x2 - y1 * y2 - z1 * z2
    x = w1 * x2 + x1 * w2 + y1 * z2 - z1 * y2
    y = w1 * y2 + y1 * w2 + z1 * x2 - x1 * z2
    z = w1 * z2 + z1 * w2 + x1 * y2 - y1 * x2
    return w, x, y, z

def q_conjugate(q):
    w, x, y, z = q
    return (w, -x, -y, -z)

def qv_mult(q1, v1):
    q2 = (0.0,) + v1
    return q_mult(q_mult(q1, q2), q_conjugate(q1))[1:]

def axisangle_to_q(v, theta):
    v = normalize(v)
    x, y, z = v
    theta /= 2
    w = cos(theta)
    x = x * sin(theta)
    y = y * sin(theta)
    z = z * sin(theta)
    return w, x, y, z

def q_to_axisangle(q):
    w, v = q[0], q[1:]
    theta = acos(w) * 2.0
    return normalize(v), theta

def q_to_mat4(q):
    w, x, y, z = q
    return numpy.array(
        [[1 - 2*y*y - 2*z*z, 2*x*y - 2*z*w, 2*x*z + 2*y*w, 0],
        [2*x*y + 2*z*w, 1 - 2*x*x - 2*z*z, 2*y*z - 2*x*w, 0],
        [2*x*z - 2*y*w, 2*y*z + 2*x*w, 1 - 2*x*x - 2*y*y, 0],
        [0, 0, 0, 1] ],'f')

def q_to_mat3(q):
    w, x, y, z = q
    return numpy.array(
        [[1 - 2*y*y - 2*z*z, 2*x*y - 2*z*w, 2*x*z + 2*y*w],
        [2*x*y + 2*z*w, 1 - 2*x*x - 2*z*z, 2*y*z - 2*x*w],
        [2*x*z - 2*y*w, 2*y*z + 2*x*w, 1 - 2*x*x - 2*y*y],
        ],'f')
```

```python
#Kalman filter smoothing:
from pyquaternion import Quaternion
import numpy
from scipy.spatial.transform import Rotation as R
import cv2

pyautogui.MINIMUM_DURATION = 0.0
pyautogui.PAUSE = 0.0
pyautogui.MINIMUM_SLEEP = 0.0
pyautogui.FAILSAFE = False

v = numpy.array([0., 1., 0.,]) # Unit vector in the front(+y)direction in this case , initialize out of the loop
it = 0

#Kalman matrices initialization
kalman = cv2.KalmanFilter(2,2)
kalman.measurementMatrix = numpy.array([[1,0],[0,1]],numpy.float32)
kalman.transitionMatrix = numpy.array([[1,0],[0,1]], numpy.float32)
kalman.processNoiseCov = numpy.array([[1,0],[0,1]], numpy.float32) * 1e-3
kalman.measurementNoiseCov = numpy.array([[1,0],[0,1]], numpy.float32) * 0.01
#initialization of starting cursor position
maxX, maxY = pyautogui.size()

kalman.statePre =  numpy.array([(maxX/2),(maxY/2)],numpy.float32)

while True:
    gyro_mot = json.loads(receivedData.recv())["mot"]
    
    if True:#gyro_mot[0] == 0:
        it+=1
#         print(gyro_mot)
        maxX, maxY = pyautogui.size()

        if it == 1:
            pyautogui.moveTo(maxX/2, maxY/2, duration=0)
            startCode = input('\n\nTo start, look at cursor at the center and type "1".\n>>> ')
            if startCode == "1":
                pass
            x_loc_old, y_loc_old = pyautogui.position()
         

        #cursor control with rotation
        scale_x = 12000/numpy.pi
        scale_y = 12000/numpy.pi
        
        #new trials with Sensor
        rot_mat= q_to_mat3(gyro_mot[2:6])
        v_out = numpy.matmul(rot_mat, v)
        x_shift_new = v_out[0]*scale_x
        y_shift_new = -v_out[2]*scale_y
        
        if it == 1:
            x_shift = 0
            y_shift = 0
        else:
            x_shift = int(x_shift_new - x_shift_old)
            y_shift = int(y_shift_new - y_shift_old)
        
        y_shift_old = y_shift_new
        x_shift_old = x_shift_new
        pyautogui.move(x_shift, y_shift)
        
#         print(gyro_mot)
#         print(x_rot, y_rot)
#         print(x_shift,y_shift)

        #save new position as Kalman measurement by adding x_shift to x_location
        x_loc_new = x_loc_old + x_shift
        y_loc_new = y_loc_old + y_shift       
        x_loc_old = x_loc_new
        y_loc_old = y_loc_new
        
        #Kalman Prediction
#         x, y = pyautogui.position()
#         a = numpy.array(pyautogui.position(),numpy.float32)

        a = numpy.array([x_loc_new,y_loc_new], numpy.float32) 
        mes = numpy.reshape(a,(2,1))
        
        correct_kalman = kalman.correct(mes)
        predict_kalman = kalman.predict()
        
        pyautogui.moveTo(predict_kalman[0],predict_kalman[1])
        
        print (kalman.statePost[0],kalman.statePost[1])
        print (kalman.statePre[0],kalman.statePre[1])
        print ('measurement:\t',mes[0],mes[1])  
        print ('correct:\t',correct_kalman[0],correct_kalman[1])
        print ('predict:\t',predict_kalman[0],predict_kalman[1])     
        print ('='*30)  
```
